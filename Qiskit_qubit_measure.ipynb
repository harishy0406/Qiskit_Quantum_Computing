# **Main Code**

```python
# Initialize |0> and |1>, measure
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator

sim = AerSimulator()

# |0> state
qc0 = QuantumCircuit(1, 1)
qc0.measure_all()
job0 = sim.run(transpile(qc0, sim), shots=512).result()
print("|0> counts:", job0.get_counts())

# |1> state (apply X)
qc1 = QuantumCircuit(1, 1)
qc1.x(0)
qc1.measure_all()
job1 = sim.run(transpile(qc1, sim), shots=512).result()
print("|1> counts:", job1.get_counts())
```

**Task 1: Verify |0⟩ and |1⟩**




```python
# |0> state
qc0 = QuantumCircuit(1, 1)
qc0.measure_all()
job0 = sim.run(transpile(qc0, sim), shots=512).result()
print("|0> counts:", job0.get_counts())

# |1> state
qc1 = QuantumCircuit(1, 1)
qc1.x(0)
qc1.measure_all()
job1 = sim.run(transpile(qc1, sim), shots=512).result()
print("|1> counts:", job1.get_counts())
```

*Are the results deterministic or random? Why?*

The results are **deterministic** because the qubits are in definite basis states (|0⟩ or |1⟩).
Measuring them always yields the same outcome (all 0s or all 1s) with 100% probability.

**Task 2: Change the number of shots (10 and 1000)**


```python
# |0> state with 10 shots
qc0 = QuantumCircuit(1, 1)
qc0.measure_all()
result_10 = sim.run(transpile(qc0, sim), shots=10).result()
print("|0> counts (10 shots):", result_10.get_counts())

# |0> state with 1000 shots
result_1000 = sim.run(transpile(qc0, sim), shots=1000).result()
print("|0> counts (1000 shots):", result_1000.get_counts())

print("-------------------------------------------------------")

# |1> state (apply X gate)
qc1 = QuantumCircuit(1, 1)
qc1.x(0)
qc1.measure_all()
result1_10 = sim.run(transpile(qc1, sim), shots=10).result()
print("|1> counts (10 shots):", result1_10.get_counts())
result1_1000 = sim.run(transpile(qc1, sim), shots=1000).result()
print("|1> counts (1000 shots):", result1_1000.get_counts())
```

*Observe the results. Do they change? Why or why not?*

The results do not change when increasing the number of shots because |0⟩ and |1⟩ are deterministic states.
Every measurement always gives the same outcome, so the counts simply scale with the number of shots (all 0’s or all 1’s).

**Task 3: Superposition with Hadamard**

```python
# |ψ> = H|0>  →  (|0> + |1>)/√2
qc = QuantumCircuit(1, 1)
qc.h(0)               # Apply Hadamard to create superposition
qc.measure_all()

# Run the simulation with 512 shots
job = sim.run(transpile(qc, sim), shots=512).result()
counts = job.get_counts()
print("Counts (Hadamard, 512 shots):", counts)
```

*What outcomes do you see? Are they deterministic or probabilistic?*

The outcomes are approximately 50% ‘0’ and 50% ‘1’.
They are probabilistic, not deterministic, because the Hadamard gate places the qubit in a superposition where both outcomes are equally likely.

**Task 4: Combine X and H gates**


```python
# Start with |0⟩, apply X then H
qc = QuantumCircuit(1, 1)
qc.x(0)           # Flip |0⟩ → |1⟩
qc.h(0)           # Apply Hadamard to |1⟩ → (|0⟩ - |1⟩)/√2
qc.measure_all()

# Run the simulation
job = sim.run(transpile(qc, sim), shots=512).result()
counts = job.get_counts()
print("Counts (X + H, 512 shots):", counts)
```

*Predict the results before running: what distribution do you expect?*

The results will be probabilistic, not deterministic — roughly half ‘0’ and half ‘1’.
This happens because the Hadamard gate acting on |1⟩ produces an equal superposition (|0⟩ − |1⟩)/√2,
giving 50% probability for each outcome when measured.
