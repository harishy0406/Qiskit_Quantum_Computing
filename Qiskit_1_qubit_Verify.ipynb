# **Main Code**

```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator

print("Qiskit AerSimulator test (Qiskit 2.x) -->1")

# Create a simple circuit: 1 qubit, 1 classical bit
qc = QuantumCircuit(1, 1)
qc.h(0)               # Apply Hadamard gate → creates superposition (|0> + |1>)/√2
qc.measure_all()      # Measure qubit and store result in classical bit

print("Circuit:")
print(qc.draw())      # Display circuit

# Run on AerSimulator
sim = AerSimulator()

# Transpile (compile) circuit for simulator backend
tqc = transpile(qc, sim)
job = sim.run(tqc, shots=1024)
result = job.result()

# Get measurement results
counts = result.get_counts()
print("Counts:", counts)
```

**Task 3: Entangled state (Bell state)**

```python
print("ANSWER:\n")

print("Qiskit AerSimulator test (Qiskit 2.x) with Entangled state ")

# Step 1: Create circuit with 2 qubits and 2 classical bits
qc1 = QuantumCircuit(2, 2)

# Step 2: Apply Hadamard to qubit 0 and CNOT (control=0, target=1)
qc1.h(0)
qc1.cx(0, 1)

# Step 3: Measure both qubits
qc1.measure_all()

# Step 4: Run on Aer simulator
tqc1 = transpile(qc1, sim)
job1 = sim.run(tqc1, shots=1024)
result1 = job1.result()
counts1 = result1.get_counts()

print(qc1.draw())
print("Measurement Results:", counts1)
```
